```python
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema

# 读取数据
file_path = 'index0-dB20Z-Stage1.csv'
data = pd.read_csv(file_path)

# 提取频率和 dB20Z 值
frequency = data['Freq [MHz]'].values
dB20Z_Rectangle2 = data['dB20(Zt(Rectangle2_T1,Rectangle2_T1)) []'].values
dB20Z_Rectangle5 = data['dB20(Zt(Rectangle5_T1,Rectangle5_T1)) []'].values

# 找到局部最小值的索引
min_indices_Rectangle2 = argrelextrema(dB20Z_Rectangle2, np.less)[0]
min_indices_Rectangle5 = argrelextrema(dB20Z_Rectangle5, np.less)[0]

# 获取第一个波谷（串联谐振点）的位置和对应的频率和 dB 值
first_valley_index_Rectangle2 = min_indices_Rectangle2[0] if len(min_indices_Rectangle2) > 0 else None
first_valley_index_Rectangle5 = min_indices_Rectangle5[0] if len(min_indices_Rectangle5) > 0 else None

resonant_frequency_Rectangle2 = frequency[first_valley_index_Rectangle2] if first_valley_index_Rectangle2 is not None else None
resonant_value_Rectangle2 = dB20Z_Rectangle2[first_valley_index_Rectangle2] if first_valley_index_Rectangle2 is not None else None

resonant_frequency_Rectangle5 = frequency[first_valley_index_Rectangle5] if first_valley_index_Rectangle5 is not None else None
resonant_value_Rectangle5 = dB20Z_Rectangle5[first_valley_index_Rectangle5] if first_valley_index_Rectangle5 is not None else None

# 输出结果
print("Rectangle2 第一个串联谐振频率:", resonant_frequency_Rectangle2, "MHz, 对应的 dB 值:", resonant_value_Rectangle2)
print("Rectangle5 第一个串联谐振频率:", resonant_frequency_Rectangle5, "MHz, 对应的 dB 值:", resonant_value_Rectangle5)
```
